```python
from os.path import join
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
## Data Import
df = pd.read_csv(join('data','abalone.csv'), names = ['Sex','Length','Diameter','Height','Whole weight','Shucked weight','Viscera weight','Shell weight','Rings'])
df.head(5)
```

```python
del df['Sex']
df
```

```python
## Some functions to explore data
print(df.corr())
print(df.dtypes)
print(df.shape)
```

```python
## scikit learn에서 제공해주는 함수 사용
from sklearn.model_selection import train_test_split
trnx, tstx, trny, tsty = train_test_split(df.iloc[:,0:7], df.iloc[:,7], test_size=0.3, random_state=123)
```

```python
from sklearn.neighbors import KNeighborsRegressor
k=5
knn_model = KNeighborsRegressor(n_neighbors=k)
knn_model.fit(X=trnx, y=trny)
```

```python
knn_pred = knn_model.predict(X=tstx)
print(knn_pred)
```

```python
## Calculation of RMSE
np.sqrt(np.mean((np.square(knn_pred - tsty))))
```

```python
plt.scatter(x=tsty, y=knn_pred)
print(np.correlate(tsty, knn_pred))
```

```python
from sklearn import metrics
print('RMSE:', np.sqrt(metrics.mean_squared_error(knn_pred, tsty)))
print('R2:', metrics.r2_score(tsty, knn_pred))
```
