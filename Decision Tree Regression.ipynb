```python
from os.path import join
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn import linear_model
from sklearn.tree import DecisionTreeRegressor

from sklearn.metrics import r2_score
from sklearn.tree import export_graphviz

%matplotlib inline
```

```python
## Data Import
df = pd.read_csv(join('data','abalone.csv'), names = ['Sex','Length','Diameter','Height','Whole weight','Shucked weight','Viscera weight','Shell weight','Rings'])
df.head(5)
```

```python
del df['Sex']
df
```

```python
## Some functions to explore data
print(df.corr())
print(df.dtypes)
print(df.shape)
```

```python
## scikit learn에서 제공해주는 함수 사용
from sklearn.model_selection import train_test_split
trnx, tstx, trny, tsty = train_test_split(df.iloc[:,0:7], df.iloc[:,7], test_size=0.3, random_state=123)
```

```python
# 결정트리 회귀 모델 생성
tr = DecisionTreeRegressor(max_depth=4)
tr.fit(trnx,trny)
```

```python
dtr_pred = tr.predict(X=tstx)
print(dtr_pred)
```

```python
from sklearn import metrics
print('RMSE:', np.sqrt(metrics.mean_squared_error(dtr_pred, tsty)))
print('R2:', metrics.r2_score(tsty, dtr_pred))
```

```python
with open("TreeRegr.dot", "w") as f:
    f = export_graphviz(tr, out_file=f)
```
